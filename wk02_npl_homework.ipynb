```python
import pandas as pd
import tiktoken
from transformers import AutoTokenizer
from kiwipiepy import Kiwi
```

```python
sentences = [
    "세조가 정인지·한확 등과 더불어 왕비를 맞아들이기를 청하다",
    "사헌부에서 중 각륙이 신분 증명서도 없이 수직한 일로 그를 논죄하기를 청하다",
    "영의정·좌의정 등이 내불당을 철거하지 않음이 좋다고 아뢰다",
    "도만호가 군사를 거느리고 순찰하게 하다",
    "사헌부가 관리의 잘못을 조사하여 논죄하다",
    "신하들이 임금께 그 일을 자세히 아뢰다",
    "간통죄를 범한 자를 엄하게 다스리다",
    "병조에서 군사의 훈련 상황을 보고하다",
    "백성의 고통이 심하니 구휼을 서두르다",
    "대신들이 조정의 일을 함께 의논하여 아뢰다"
]
```

```python
# 1) BPE tokenizer (GPT 계열)
bpe_tokenizer = tiktoken.encoding_for_model("gpt-4")

# 2) SentencePiece 계열 tokenizer
sp_tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")

# 3) 형태소 분석기
kiwi = Kiwi()
```

```python
def tokenize_bpe(text):
    token_ids = bpe_tokenizer.encode(text)
    tokens = [bpe_tokenizer.decode([tid]) for tid in token_ids]
    return tokens, len(token_ids)

def tokenize_sp(text):
    tokens = sp_tokenizer.tokenize(text)
    return tokens, len(tokens)

def tokenize_kiwi(text):
    tokens = [token.form for token in kiwi.tokenize(text)]
    return tokens, len(tokens)
```

```python
test_text = "영의정이 그 일을 아뢰다"

print("BPE:", tokenize_bpe(test_text))
print("SentencePiece:", tokenize_sp(test_text))
print("Kiwi:", tokenize_kiwi(test_text))
```

```python
def tokenize_bpe(text):
    token_ids = bpe_tokenizer.encode(text)
    tokens = [bpe_tokenizer.decode([tid]) for tid in token_ids]
    return tokens, len(token_ids)

def tokenize_sp(text):
    tokens = sp_tokenizer.tokenize(text)
    return tokens, len(tokens)

def tokenize_kiwi(text):
    tokens = [token.form for token in kiwi.tokenize(text)]
    return tokens, len(tokens)
```

```python
rows = []

for sent in sentences:
    bpe_tokens, bpe_count = tokenize_bpe(sent)
    sp_tokens, sp_count = tokenize_sp(sent)
    kiwi_tokens, kiwi_count = tokenize_kiwi(sent)

    rows.append({
        "sentence": sent,
        "BPE_tokens": bpe_tokens,
        "BPE_count": bpe_count,
        "SentencePiece_tokens": sp_tokens,
        "SentencePiece_count": sp_count,
        "Kiwi_tokens": kiwi_tokens,
        "Kiwi_count": kiwi_count
    })

df = pd.DataFrame(rows)
df
```

```python
df[["BPE_count", "SentencePiece_count", "Kiwi_count"]].mean()
```

```python
avg_df = df[["BPE_count", "SentencePiece_count", "Kiwi_count"]].mean().reset_index()
avg_df.columns = ["tokenizer", "avg_token_count"]
avg_df
```

```python
test_words = ["청하다", "아뢰다", "영의정", "도만호", "사헌부", "논죄", "간통죄", "순찰"]

for word in test_words:
    print("=" * 50)
    print("WORD:", word)
    print("BPE:", tokenize_bpe(word))
    print("SentencePiece:", tokenize_sp(word))
    print("Kiwi:", tokenize_kiwi(word))
    print()
```

```python
focus_sentences = [
    "영의정·좌의정 등이 내불당을 철거하지 않음이 좋다고 아뢰다",
    "사헌부에서 중 각륙이 신분 증명서도 없이 수직한 일로 그를 논죄하기를 청하다",
    "도만호가 군사를 거느리고 순찰하게 하다"
]

for sent in focus_sentences:
    print("=" * 80)
    print("SENTENCE:", sent)
    print("BPE:", tokenize_bpe(sent))
    print("SentencePiece:", tokenize_sp(sent))
    print("Kiwi:", tokenize_kiwi(sent))
    print()
```
